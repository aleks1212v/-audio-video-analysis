```python
import pandas as pd
import numpy as np
import random
import os
import soundfile as sf
import librosa.util
import librosa.display
from pydub import AudioSegment, effects
from pydub.silence import split_on_silence
import zipfile
import shutil
from shutil import make_archive
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
#сохранение csv
def mfcc_plot(path, name):
    num_of_song = 30
    try:
        x, sr = sf.read(os.path.join(path, name))
        len_frame = x.shape[0] // num_of_song #220500
        frame_list = list()
        for i in range(num_of_song):
            frame_list.append(x[i * len_frame : (i + 1) * len_frame])
        mfccs_list = list()
        for frame in frame_list:
            mfccs = librosa.feature.mfcc(frame, sr=sr, n_mfcc = 15, dct_type = 2, norm = 'ortho')
            mfccs_list.append(mfccs)
            #print(mfccs.shape)
            #print(x.shape)
            #print(sr)
            # Отображение MFCC:
            #plt.figure(figsize=(15, 7))
            #librosa.display.specshow(mfccs, sr=sr, x_axis='time')
            #plt.title(name)
            #plt.show()
        df_mfccs = pd.DataFrame()
        for mfccs in mfccs_list:
            df = pd.DataFrame(mfccs).transpose()
            df_mfccs = pd.concat([df_mfccs, df], ignore_index = True)
        if df_mfccs.shape == (12930, 15):
            print(f"Запись файла {Path(name).stem + '.csv'}, размер {df_mfccs.shape}")
            df_mfccs.to_csv(os.path.join(path, Path(name).stem + '.csv'), index = None)
        else:
            print(f'ЧТО-ТО НЕ ТАК С ФАЙЛОМ {name}')
    except Exception as err:
        print(f'Не удалось создать файл {name}, {err}')

if __name__ == '__main__':
    dirpath = os.getcwd()
    for new_root, dirs, files in os.walk(dirpath):
        for file in files:
            if Path(file).suffix == '.wav':
                mfcc_plot(new_root, file)
```
